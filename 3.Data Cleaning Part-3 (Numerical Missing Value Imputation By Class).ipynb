```python
#import libaries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Dataset 
dataset_path = r"https://drive.google.com/uc?export=download&id=1BiGZSedP4BIIuTbVTBodOhVgFImaz08c"
df = pd.read_csv(dataset_path)
```

```python
df.shape
```

```python
#To show all column and row
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
#Missing value with percentages
df_null_per = df.isnull().sum()/df.shape[0]*100
df_null_col = df_null_per[df_null_per>20].keys()
df_null_col
#drop missing values from upper 20 percentage
df_drop_null = df.drop(columns=df_null_col)
df_drop_null.shape
```

```python
df_drop_null.isnull().sum()
```

```python
#Find numeric value column from dataset
df_num = df_drop_null.select_dtypes(include=['int64','float64'])
```

```python
#find null/missig column
df_num.isnull().sum()
```

```python
df_num_null_per = df_num.isnull().sum()/df_num.shape[0]*100
df_num_null = df_num_null_per[df_num_null_per>0].keys()
df_num_null
```

```python
#For 'LotFrontage', 'LotConfig' is accurate class
df['LotConfig'].unique()
```

```python
#first select the attribute/column and check with class. then missing column
df1_class = df[df.loc[:,'LotConfig'] == 'Inside']['LotFrontage']

#then use replace function
df1_mean = df1_class.replace(np.nan,df1_class.mean())

df1_mean

#df[df.loc[:,'LotConfig'] == "Inside"]["LotFrontage"].replace(np.nan,df[df.loc[:,'LotConfig'] == "Inside"]["LotFrontage"].mean())
```

```python
#Find mean by class for one attribute
df_copy = df.copy()
for var_class in (df['LotConfig'].unique()):
    df_copy.update(df[df.loc[:,'LotConfig'] == var_class]["LotFrontage"].replace(np.nan,df[df.loc[:,'LotConfig'] == var_class]["LotFrontage"].mean()))
    
```

```python
df_copy.isnull().sum()
```

# missing value fillup by Mean

```python
#fillup missing value
df_copy = df.copy()
num_null = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_class = ['LotConfig','MasVnrType','GarageType' ]
for var_num, var_cat in zip(num_null,cat_class):
    for var_class in (df[var_cat].unique()):
        df_copy.update(df[df.loc[:,var_cat] == var_class][var_num].replace(np.nan,df[df.loc[:,var_cat] == var_class][var_num].mean()))
    
```

```python
#Selecting class has missing value thats why it is not working
df_copy[df_num_null].isnull().sum()
```

```python
df_copy_mean = df.copy()
num_null = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_class = ['LotConfig','Exterior2nd','KitchenQual' ]
for var_num, var_cat in zip(num_null,cat_class):
    for var_class in (df[var_cat].unique()):
        df_copy_mean.update(df[df.loc[:,var_cat] == var_class][var_num].replace(np.nan,df[df.loc[:,var_cat] == var_class][var_num].mean()))
    
```

```python
#Selecting class has no missing value now thats why it is working
df_copy_mean[df_num_null].isnull().sum()
```

# Data Distribution

# Data Distribution by Mean

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(df_num_null):
    plt.subplot(2,2,i+1)
    sns.distplot(df_num[var], bins=20, kde_kws={'linewidth':5, 'color':'red'}, label="Original",)
    sns.distplot(df_copy_mean[var], bins=20, kde_kws={'linewidth':5, 'color':'green'}, label="mean",)
    plt.legend()
```

# missing value fillup by Median

```python
df_copy_median = df.copy()
num_null = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_class = ['LotConfig','Exterior2nd','KitchenQual' ]
for var_num, var_cat in zip(num_null,cat_class):
    for var_class in (df[var_cat].unique()):
        df_copy_median.update(df[df.loc[:,var_cat] == var_class][var_num].replace(np.nan,df[df.loc[:,var_cat] == var_class][var_num].median()))
    
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(df_num_null):
    plt.subplot(2,2,i+1)
    sns.distplot(df_num[var], bins=20, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(df_copy_mean[var], bins=20, kde_kws={'linewidth':5, 'color':'green'}, label="mean",)
    sns.distplot(df_copy_median[var], bins=20, kde_kws={'linewidth':3, 'color':'k'}, label="mean",)
    plt.legend()
```

```python
#Check outlier with boxplot
for i, var in enumerate(df_num_null):
    plt.figure(figsize=(10,10))
    plt.subplot(3,1,1)
    sns.boxplot(df_num[var])
    plt.subplot(3,1,2)
    sns.boxplot(df_copy_mean[var])
    plt.subplot(3,1,3)
    sns.boxplot(df_copy_median[var])
```
