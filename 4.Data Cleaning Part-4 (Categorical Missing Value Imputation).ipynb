```python
#import libaries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Dataset 
dataset_path = r"https://drive.google.com/uc?export=download&id=1BiGZSedP4BIIuTbVTBodOhVgFImaz08c"
df = pd.read_csv(dataset_path)
```

```python
df.shape
```

```python
#To show all column and row
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
#Missing value with percentages
df_null_per = df.isnull().sum()/df.shape[0]*100
df_null_col = df_null_per[df_null_per>20].keys()
df_null_col
#drop missing values from upper 20 percentage
df_drop_null = df.drop(columns=df_null_col)
df_drop_null.shape
```

```python
df_drop_null.isnull().sum()
```

```python
#Find object value column from dataset
df_obj = df_drop_null.select_dtypes(include=['object'])
df_obj.head()
```

```python
df_obj.isnull().sum()
```

```python
#Find missing object column
df_per_obj = df_obj.isnull().mean()*100
df_null_obj = df_per_obj[df_per_obj>0].keys()
df_null_obj
```

```python
df_obj['MasVnrType'].value_counts()
```

```python
#Add manual object 'Missing' and check
df_obj['MasVnrType'].fillna('Missing').value_counts()
```

```python
df_obj['MasVnrType'].mode()
```

```python
#Add maximum value using mode()
df_obj['MasVnrType'].fillna(df_obj['MasVnrType'].mode()[0]).value_counts()
```

```python
df_obj_copy = df_obj.copy()

for var in df_null_obj:
    df_obj_copy[var].fillna(df_obj[var].mode()[0],inplace=True)
    print(var,"=",df_obj[var].mode()[0])
```

```python
df_obj_copy.isnull().sum()
```

```python
plt.figure(figsize=(16,9))
for i,var in enumerate(df_null_obj):
    plt.subplot(4,3,i+1)
    plt.hist(df_obj_copy[var],label="Impute")
    plt.hist(df_obj[var].dropna(),label="Original")
    plt.legend()
```

```python
#Update the dataset and again delete above 20% missing column
df.update(df_obj_copy)
df.drop(columns=df_null_col,inplace=True)
```
