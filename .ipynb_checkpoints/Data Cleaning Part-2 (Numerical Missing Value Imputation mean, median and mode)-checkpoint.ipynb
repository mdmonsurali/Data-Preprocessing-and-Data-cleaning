```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Dataset 
dataset_path = r"https://drive.google.com/uc?export=download&id=1BiGZSedP4BIIuTbVTBodOhVgFImaz08c"
df = pd.read_csv(dataset_path)
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
null_ver_per = df.isnull().sum()/df.shape[0]*100
null_ver_per
```

```python
drop_col = null_ver_per[null_ver_per>20].keys()
drop_col
```

```python
df_drop_col = df.drop(columns=drop_col)
df_drop_col.shape
```

```python
df_num = df_drop_col.select_dtypes(include=['int64','float64'])
df_num.head(5)
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df_num.isnull())
```

```python
df_num_null = df_num.isnull().sum()

df_null_col= df_num_null[df_num_null>0].keys()
df_null_col
```

```python
#Note when distribution is normal distribution like 1st pic then use mean, median or mode. Otherside when median
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(df_null_col):
    plt.subplot(2,2,i+1)
    sns.distplot(df_num[var], bins=20, kde_kws={'linewidth':5, 'color':'#DC143C'})
```

```python
df_num_mean = df_num.fillna(df_num.mean())
df_num_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(df_null_col):
    plt.subplot(2,2,i+1)
    sns.distplot(df_num[var], bins=20, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(df_num_mean[var], bins=20, kde_kws={'linewidth':8, 'color':'green'}, label="mean",)
    plt.legend()
```

```python
#Apply median

df_num_median = df_num.fillna(df_num.median())
df_num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(df_null_col):
    plt.subplot(2,2,i+1)
    sns.distplot(df_num[var], bins=20, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(df_num_median[var], bins=20, kde_kws={'linewidth':8, 'color':'green'}, label="median",)
    plt.legend()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(df_null_col):
    plt.subplot(2,2,i+1)
    sns.distplot(df_num[var], bins=20,hist=False, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(df_num_mean[var], bins=20,hist=False, kde_kws={'linewidth':5, 'color':'green'},label="Mean",)
    sns.distplot(df_num_median[var], bins=20,hist=False, kde_kws={'linewidth':3, 'color':'k'},label="Median",)
    plt.legend()
```

```python
#check outlier
for i, var in enumerate(df_null_col):
    plt.figure(figsize=(10,10))
    plt.subplot(3,1,1)
    sns.boxplot(df_num[var])
    plt.subplot(3,1,2)
    sns.boxplot(df_num_mean[var])
    plt.subplot(3,1,3)
    sns.boxplot(df_num_median[var])
```

```python
#compare
df_concat = pd.concat([df_num[df_null_col],df_num_mean[df_null_col],df_num_median[df_null_col]], axis=1)
df_concat[df_concat.isnull().any(axis=1)]
```
